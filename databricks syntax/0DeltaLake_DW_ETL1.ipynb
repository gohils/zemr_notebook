```
%sql
CREATE DATABASE IF NOT EXISTS cdw_prod;
USE cdw_prod;
```

```
# create delta table multiple table based on datalake file location
src_name = ['DIM_Customer','DIM_Product','DIM_Promotion_master','FACT_Sales']
from pyspark.sql.functions import *
for tables in src_name:
    loc = f"/mnt/cdw/Transformed_Data/delta_lake/{tables}"
    spark.sql(f"DROP TABLE IF EXISTS cdw_prod.{tables}")
#     spark.sql(f"CREATE TABLE IF NOT EXISTS cdw_prod.{tables} USING PARQUET LOCATION'{loc}'")
    spark.sql(f"CREATE TABLE IF NOT EXISTS cdw_prod.{tables}  USING DELTA LOCATION '{loc}'")
```

```
%sql
-- Aggregating query with joins
SELECT C.CountryName, P.ProductSubcategoryKey,
 d.Year,
 SUM(f.SalesAmount) AS Sales
FROM cdw_prod.fact_sales AS f
 left outer JOIN cdw_prod.dim_customer AS C
  ON f.CustomerKey = C.CustomerKey
 LEFT OUTER JOIN cdw_prod.dim_product AS P
  ON f.ProductKey = p.ProductKey
 LEFT OUTER JOIN cdw_prod.dim_date AS d
  ON f.OrderDateKey = d.DateKey
GROUP BY C.CountryName, P.ProductSubcategoryKey, d.Year
ORDER BY C.CountryName, P.ProductSubcategoryKey, d.Year
;
```
